```python
# 문제데이터, 정답데이터 분리
X = data.drop("poisonous", axis=1)
y = data['poisonous']
```

### 3.데이터전처리
- 문자열 데이터 -> 수치화 : 인코딩
- 인코딩 방법 2가지
    - 라벨인코딩 : 값의 크고 작음의 의미가 유용한 경우 활용
    - 원핫인코딩 : 값의 크고 작음의 의미가 없을 때 활용

```python
# 원핫인코딩
# 컬럼의 unique 개수만큼 컬럼을 추가한 뒤, 1 or 0을 채워주는 방식
# pd.get_dummies(문제데이터)
X_one_hot = pd.get_dummies(X)
X_one_hot
# 총 117ea의 컬럼으로 확장됨!
```

```python
# 라벨인코딩
# 각 문자열에 숫자를 매핑하는 방법
X['habitat'].unique()
```

```python
# mapping할 클래스값과 숫자값을 짝지어줌
# 숫자가 클수록 가중치↑
hb_dict = {
    'u':1, 
    'g':2, 
    'm':3, 
    'd':4, 
    'p':5, 
    'w':6, 
    'l':7
}

# 특정대상을 딕셔너리에 넣고, map() 매개변수로 지정
X['habitat'].map(hb_dict)
```

```python
y_map = y.map({'p':1,'e':0})
```

```python
# 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y ,test_size=0.3, 
                                                    random_state=2024, 
                                                    stratify=y)

# 역할
# 1. 데이터 분리
# 2. 랜덤 샘플링 → 랜덤 스테이트(랜덤키 고정)
```

```python
# 크기 확인!!!!(문제와 정답 데이터 갯수, 특성의 갯수 확인 꼭 해줘야 함!)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### 6.모델학습

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

### 7.모델평가

```python
tree.score(X_test, y_test)
# 과대적합 -> 100
```

### 특성 선택

- 의사결정나무는 특성의 중요도가 노은 특성을 상단에 우선 배치
- 지도학습 모델에서는 데이터들의 특성의 중요도를 출력
- 특성들 중 정답에 가장 큰 영향을 미칠수록 큰 수치
- 각 특성들은 0~1 사이의 수치를 가지며 특성중요도의 총 합은 1
