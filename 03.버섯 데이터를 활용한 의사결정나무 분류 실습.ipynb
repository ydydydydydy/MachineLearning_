### 3. 데이터 전처리
- 문자열 데이터 -> 수치화 : 인코딩
- 인코딩 방법 2가지
    - 라벨인코딩 : 값의 크고 작음에 의미가 유용한 경우 활용
    - 원핫인코딩 : 값의 크고 작음에 의미가 없을 때 활용

```python
#원핫인코딩
#컬럼의 unique 개수만큼 컬럼을 추가한 뒤 1 또는 0으로 채워주는 방식
X_one_hot = pd.get_dummies(X)

#컬럼들이 쪼개지므로 무려 117ea로 늘어난다.
```

```python
#라벨인코딩
#각 문자열에 숫자를 매핑하는 방법
X['habitat'].unique()
```

```python
y
```

```python
#mapping할 클래스값(key)과 숫자값(value)으로 짝지어 딕셔너리 형태로 만듬
hb_dict = {
    'u':1, 
    'g':2, 
    'm':3, 
    'd':4, 
    'p':5, 
    'w':6, 
    'l':7
} 

#보통 중요한 데이터에 높은 숫자를 주는 방식으로 가중치를 둔다
#특정 대상을 딕셔너리에 넣고 map() 매개변수로 지정
X['habitat'].map(hb_dict) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, 
                                                    test_size=0.3, 
                                                    random_state=2024, 
                                                    stratify=y)
#train_test_split()의 기능
#1. 데이터 분리
#2. 랜덤샘플링 -> random 키를 고정해 줌.(고정하지 않으면 하이퍼 파라미터 조정할 때 문제 생김)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### 5. 모델 선택 및 하이퍼파라미터 조정

### 6. 모델 학습

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
```

### 7. 모델 평가

```python
#score만 쓸 거면 굳이 predict()안 써도 된다.

tree.score(X_test,y_test)

#100%가 나왔으므로 좋은 모델? NO, 과대적합일 가능성이 높다.
#순수 노드가 너무 많다는 건 결국 정해진 루트로만 간다는 의미이므로
#다른 테스트셋에서는 예측값을 잘 맞추지 못할 것
#기본적으로 의사결정 나무는 순수 노드만 남을 때까지 분류한다.
#제어할 필요성
```

#### 하이퍼파라미터 조정을 통한 과적합 제어

```python
tree2 = DecisionTreeClassifier(max_depth=3) 
#최대 depth를 3으로 제한하여 순수노드만 남는 과대적합 방지.
tree2.fit(X_train,y_train)
```

```python
tree2.score(X_test,y_test)
```

### 교차검증
- 모델의 일반화 여부 확인을 위하여
- 학습-평가 데이터 나누기를 여러 번 반복하여 일반화 error를 평가하는 방법
- 테스트 데이터는 최종 평가용으로 남겨두고 학습용 데이터 내에서 자체적으로 검증하는 것


```python
#k-fold 교차검증 
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(tree2,X_train,y_train, 
                cv=5) #만들어준 모델과 트레인셋,k값을 설정해 준다.

#전반적으로 수치가 높다 -> 일반화 가능성 높음!
```

```python
cross_val_score(tree2,X_train,y_train, 
                cv=5).mean() 
```

```python
#depth가 설정되어 있지 않아 과대적합일 확률이 높아보이는 
#첫번째 모델도 교차검증해 보자.
cross_val_score(tree,X_train,y_train,cv=5)
#5개의 값 중 4개가 100% => 과대적합은 아님. 과대적합이려면 전부 100%여야.
#특성 분리가 잘 되어있는 데이터로 학습하였기 때문.
#즉 식용과 독버섯의 차이가 너무 명확하다는 의미.
```

### 특성선택
- 의사결정나무가 의사결정을 할 때 선택하는 질문(특성 선택) 우선순위 확인
- 의사결정나무는 특성의 중요도가 높은 특성을 상단에 우선배치
- 지도학습 모델에서는 데이터들의 특성의 중요도를 출력
- 특성들 중 정답에 가장 큰 영향을 미칠수록 큰 수치
- 각 특성들은 0~1 사이의 수치

```python
#전체 컬럼명 확인
pd.set_option('display.max_seq_items', None) 
#컬럼명 생략하지 말고 전부 다 보여주라는 것
X_one_hot.columns
```

```python
#특성 중요도 확인
fi = tree.feature_importances_
```

```python
#컬럼 전체 보기(이건 항상 보고자 하는 것 위쪽에 설정해 주어야 한다)
pd.set_option('display.max_rows',None)

#DataFrame으로 만들기
fi_df = pd.DataFrame(fi, index=X_one_hot.columns, columns=['특성 중요도'])

#내림차순으로 정리
fi_df.sort_values('특성 중요도', ascending = False)

#결과
#117개의 컬럼 중에 실제 결과에 영향을 미치는 건 13개 뿐.
#필요없는 컬럼들은 삭제해버리는 게 낫다.
```
