### 문제정의
- 와인정보를 토대로 화이트/레드 여부 구분
- 분류실습(class : 2개)

### 데이터 수집
- 0:red, 1:white

```python
import pandas as pd

red_wine = pd.read_csv('data/winequality-red.csv', sep=';')
white_wine = pd.read_csv('data/winequality-white.csv', sep=';') #sep : 구분자
```

```python
#정답데이터 만들어주기
red_wine['color'] = 0
white_wine['color'] = 1
```

```python
red_wine.head()
```

```python
white_wine.head()
```

```python
#데이터 프레임 병합
wine = pd.concat([red_wine,white_wine]) #횡으로 병합하려면 axis 값 수정하면 됨.

#인덱스 번호 겹치는 문제 해결 : 재정렬
wine.reset_index(drop=True, inplace=True)
```

```python
wine
```

```python
#데이터 확인
wine.info()
#입력 특성 12개
#정답데이터 color 
```

### 전처리
- 필요없음

### 데이터 분리
- X, y
- 훈련용, 평가용

```python
X = wine.drop(columns='color')#혹은 'color', axis = 1로 해도 된다.
y = wine['color']
```

```python
#훈련용, 평가용 분리 > 분리도구 사용하기 7:3, random_state = 19
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7, random_state=19, stratify=y)

#train_size에 그냥 정수를 넣으면? 비율이 아닌 갯수가 된다. 즉 train_size = 7이라고 하면 7개만 
#train이 됨.
```

```python
#데이터 크기 확인
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train
```

### 모델 선택
- Logistic Regression 실습

```python
from sklearn.linear_model import LogisticRegression #선형분류모델!
#regression이라는 '회귀' 단어가 들어가지만 분류 모델이다!
```

```python
lr_model = LogisticRegression()
```

```python
#학습
lr_model.fit(X_train,y_train)
```

```python
#교차검증
from sklearn.model_selection import cross_val_score
cross_val_score(lr_model, X_train,y_train, cv=5)
```

```python
#평가
lr_model.score(X_test,y_test)

#약 0.98 정확도가 나오는 것을 확인
```

### 학습기준(loss,error,cost)과 평가지표(metrics)
- 학습기준(loss)
    - 분류 : 교차 엔트로피 오차(Cross Entropy Error)
    - 회귀 : MSE
- 평가지표(metrics)
    - 분류 : 정확도(accuracy), 재현율(recall), 정밀도(precision), F1-Score
    - 회귀 : MSE,RMSE,MAE,R2 Score

### SVM 모델(Linear Support Vector Machine)
    - 마진(클래스를 구분하는 데이터들간의 거리, 여백)이 최대가 되는 직선을 그림.



- 결정경계와 가장 가까운 데이터와의 거리를 Margin이라고 함.
- 이 마진이 최대화가 되는 선형 분류 결정경계를 찾는 것이 SVM의 원리
    - 고로 이 셋 중에서는 2가 가장 최적의 결정경계
    


```python
from sklearn.svm import LinearSVC
```

```python
# 모델 생성
svc_model = LinearSVC() #SVC란 Support Vector Machine 중에서 Classifier
```

```python
# 학습
svc_model.fit(X_train,y_train)
```

```python
#교차 검증
cross_val_score(svc_model,X_train,y_train,cv=5)

#교차검증에서 과적합이 아님을 판단하는 기준 : 교차검증된 값들이 비슷한 수준이라면 일반화가
#잘 되었다고 판단할 수 있다.
```

```python
#평가
svc_model.score(X_test,y_test)
```

### 다시 한 번 학습기준(error,loss,cost)과 평가지표(metrics) 정리

[학습시 사용하는 오차]
- 회귀
   - MSE
- 분류
   - 크로스엔트로피(cross entropy)

[평가지표 (metrics)]
- 회귀
   - MSE (평균제곱오차)
   - RMSE
   - MAE
   - R2 score
- 분류
   - accuracy (정확도): 전체데이터에서 맞춘 비율
   - recall(재현율): 실제양성 중에서 예측양성 비율
   - precision(정밀도): 예측양성 중에서 실제 양성 비율
   - f1-score (조화평균): 정밀도와 재현율의 조화평균(2*(recall * precision))/recall + precision)
